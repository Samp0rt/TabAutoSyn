```python
import pandas as pd
from tabautosyn import TabAutoSyn
import os
```

```python
df_path_train = '/home/user/Desktop/AutoML/TabAutoSyn/datasets_test/train'
df_path_test = '/home/user/Desktop/AutoML/TabAutoSyn/datasets_test/test'

data_info = pd.read_csv('/home/user/Desktop/AutoML/TabAutoSyn/datasets_test/data_info.csv', index_col='df_name')
```

```python
ctgan_results = {}
ddpm_results = {}
dpgan_results = {}

models = ['universal', 'privacy', 'ml']
models_map = {
    'universal': 'ctgan',
    'ml': 'ddpm',
    'privacy': 'dpgan'}

for task in models:
    print(f'=== Generating results for {models_map[task]} ===\n')
    for df in os.listdir(df_path_train):
        print(f'--- Starting synthetics generation for {df} ---')
        train_path = os.path.join(df_path_train, df)
        test_path = os.path.join(df_path_test, df)

        test_df = pd.read_csv(test_path)
        len_syn = len(test_df)

        target_column = data_info.loc[df.split('.')[0], 'target_name']

        tabsyn = TabAutoSyn(model='task_specific', task=task, verbose=True)
        syn_df = tabsyn.generate(train_data_path=train_path, sep=',', n_samples=len_syn, target_column=target_column)
        
        dict_name = f'{models_map[task]}_results'
        dict_name[df.split('.')[0]] = syn_df
```

```python
ctgan_results = {'breast_cancer': pd.DataFrame({'col1': [1, 2], 'col2': [3, 4]})}
ddpm_results = {'iris': pd.DataFrame()}
dpgan_results = {'wine': pd.DataFrame()}

results = {
    'ctgan': ctgan_results,
    'ddpm': ddpm_results, 
    'dpgan': dpgan_results}

for model_results_key in results.keys():
    key = model_results_key
    for df_name, df in results[key].items():
        df.to_csv(f'/home/user/Desktop/AutoML/TabAutoSyn/results_exp/{key}/{df_name}.csv')

print('ctgan', os.listdir('/home/user/Desktop/AutoML/TabAutoSyn/results_exp/ctgan'))
print('ddpm', os.listdir('/home/user/Desktop/AutoML/TabAutoSyn/results_exp/ddpm'))
print('dpgan', os.listdir('/home/user/Desktop/AutoML/TabAutoSyn/results_exp/dpgan'))
```
